```python
def num_points(a,b,p):
  c = 0
  #counter
  if sp.isprime(p) != True:
    raise ValueError("p is not prime")
    #if p is not prime, raise error
  else:
    for x in range(p):
      #for x less than p...
      for y in range(p):
        #...and for y less than p...
        k = y**2 - x**3 - a*x - b
        if k%p == 0:
          c = c+1
          #...if k is a multiple of p, add one to the counter...
        else:
          c = c
          #...if not, do nothing
  return c
```
